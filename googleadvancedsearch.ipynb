```python
#import statements
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
import time
```

```python
#initialize webdriver
driver = webdriver.Chrome()
driver.get("https://www.google.com/advanced_search")
WebDriverWait(driver, 2)
```

```python
#Google Advanced Search
search = driver.find_element(By.CLASS_NAME, 'jfk-textinput')
search.click()
query = "windows, texas"
search.send_keys(query)
button = driver.find_element(By.XPATH, '//*[@id="s1zaZb"]/div[5]/div[9]/div[2]/input[2]')
button.click()

#press more buisnesses button
moreBuisnesses = driver.find_element(By.XPATH, '/html/body/div[7]/div/div[11]/div[1]/div[2]/div/div/div/div/div[2]/div[1]/div[2]/g-more-link/a/div')
moreBuisnesses.click()
```

```python
#scrape companies
mydict = {"Distributor Name": [], "Phone": [], "Website Url": []}
while(True):
    WebDriverWait(driver, 2)
    searchResults = driver.find_elements(By.CLASS_NAME, 'VkpGBb') 
    for i in searchResults:
        WebDriverWait(i, 2)
        #phone numbers
        try:
            fs = i.find_element(By.XPATH, './div/a/div/div/div[4]')
            fullstr = fs.text
            phoneNumber = fullstr[-14:-1]
            if(fullstr[-14:-13] != "("):
                phoneNumber = "N/A"
            mydict["Phone"].append(phoneNumber)
        except:
            mydict["Phone"].append("N/A")
        #distributor names
        try:
            companies = i.find_element(By.XPATH, './div/a/div/div/div[1]/span')
            mydict["Distributor Name"].append(companies.text)
        except: 
            mydict["Distributor Name"].append("N/A")
        #website urls
        try:
            websites = i.find_element(By.XPATH, './a[1]')
            mydict["Website Url"].append(websites.get_attribute('href'))
        except: 
            mydict["Website Url"].append("N/A")
    nextButton = driver.find_element(By.XPATH, "//td[@class='d6cvqb BBwThe']/a")
    WebDriverWait(i, 5)
    try:
        nextButton.click()
    except:
        break
columnHeaders = ["Distributor Name", "Phone", "Data Source url"]
df = pd.DataFrame(mydict).T
df
```

```python
fileTitle = query + ".csv"
df.to_csv(fileTitle)
```
